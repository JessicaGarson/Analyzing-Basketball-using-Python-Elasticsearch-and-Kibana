```python
from nba_api.stats.static import teams
from nba_api.stats.endpoints import leaguegamefinder
from elasticsearch import Elasticsearch, helpers
from getpass import getpass
from io import StringIO
import pandas as pd
```

```python
nba_teams = teams.get_teams()
```

```python
nba_teams
```

```python
celtics = [team for team in nba_teams if team['abbreviation'] == 'BOS'][0]
```

```python
timberwolves = [team for team in nba_teams if team['abbreviation'] == 'MIN'][0]
```

```python
celtics_id = celtics['id']
```

```python
timberwolves_id = timberwolves['id']
```

```python
celtics_gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=celtics_id)
celtics_games = celtics_gamefinder.get_data_frames()[0]
celtics_games.head()
```

```python
timberwolves_gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=timberwolves_id)
timberwolves_games = timberwolves_gamefinder.get_data_frames()[0]
timberwolves_games.head()
```

```python
celtics_current_season = celtics_games.loc[(celtics_games['GAME_DATE'] >= '2023-10-24') & (celtics_games['GAME_DATE'] <= '2024-06-20')]
celtics_current_season
```

```python
celtics_current_season.isnull().values.any()
```

```python
timberwolves_current_season = timberwolves_games.loc[(timberwolves_games['GAME_DATE'] >= '2023-10-24') & (timberwolves_games['GAME_DATE'] <= '2024-06-20')]
timberwolves_current_season
```

```python
timberwolves_current_season.isnull().values.any()
```

```python
celtics_current_season_copy = celtics_current_season.copy()

celtics_current_season_copy.fillna(0, inplace=True)
```

```python
es = Elasticsearch(
    getpass("Host: "),
    api_key=getpass("API Key: "),
)
```

```python
celtics_index_name = "osn_prac_demo_celtics"
```

```python
timberwolves_index_name = "osn_prac_demo_timberwolves"
```

```python
es.indices.create(index=celtics_index_name)
```

```python
es.indices.create(index=timberwolves_index_name)
```

```python
def doc_generator(df, timeframe):
    for index, document in df.iterrows():
        yield {
            "_index": timeframe, 
            "_id": f"{document['GAME_ID']}",
            "_source": document.to_dict(),
        }
```

```python
helpers.bulk(es, doc_generator(celtics_current_season_copy, celtics_index_name))
```

```python
helpers.bulk(es, doc_generator(timberwolves_current_season, timberwolves_index_name))
```

```python
search_query = {
    "query": {
        "match": {
            "WL": "W"
        }
    }
}
```

```python
celtics_games_won = es.count(index=celtics_index_name, body=search_query)
timberwolves_games_won = es.count(index=timberwolves_index_name, body=search_query)
```

```python
celtics_games_won
```

```python
timberwolves_games_won
```

```python
print(f"The Celtics won {celtics_games_won['count']} games this season so far.")
```

```python
print(f"The Timberwolves won {timberwolves_games_won['count']} games this season so far.")
```

```python
streak_query = {
  "size": 1000,  
  "sort": [
    {
      "GAME_DATE": {
        "order": "asc"
      }
    }
  ],
  "_source": ["GAME_DATE", "WL"]
}
```

```python
celtics_streak_search = es.search(index=celtics_index_name, body=streak_query)
```

```python
timberwolves_strek_search = es.search(index=timberwolves_index_name, body=streak_query)
```

```python
cs = [hit['_source'] for hit in celtics_streak_search['hits']['hits']]
```

```python
ts = [hit['_source'] for hit in timberwolves_strek_search['hits']['hits']]
```

```python
# celtics top 5 streaks
```

```python
streaks = []
current_streak = 1
for i in range(1, len(cs)):
    if cs[i]['WL'] == cs[i-1]['WL']:
        current_streak += 1
    else:
        streaks.append((cs[i-1]['WL'], current_streak))
        current_streak = 1


streaks.append((cs[-1]['WL'], current_streak))
top_streaks = sorted(streaks, key=lambda x: x[1], reverse=True)[:5]
print(top_streaks)
```

```python
# timberwolves top 5 streaks
```

```python
streaks = []
current_streak = 1
for i in range(1, len(ts)):
    if ts[i]['WL'] == ts[i-1]['WL']:
        current_streak += 1
    else:
        streaks.append((ts[i-1]['WL'], current_streak))
        current_streak = 1


streaks.append((ts[-1]['WL'], current_streak))
top_streaks = sorted(streaks, key=lambda x: x[1], reverse=True)[:5]
print(top_streaks)
```
